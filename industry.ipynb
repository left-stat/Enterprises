```python
df = pd.read_excel('C:\\Users\\yraz\\Enterprises\\data\\data.xlsx')
print("Количество компаний в списке: ", len(df), ' компаний')
print('Первые 5 компаний\n')
df.head()
```

```python
print('Распределение компаний по отраслям с сортировкой по размеру выручки\n')
print('Столбец v - выручка за 2017 год')
print('Столбцы p - выручка по годам')
print('Столбцы p - количество сотрудников по годам')
industry = df.groupby('industry').sum()
industry = industry.sort_values('v', ascending=False)
industry
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(14, 8))
plt.title('\nРаспределение отраслей по выручке за 2017 год (в рублях)\n', fontsize=16)
plt.ylabel('\nВыручка компаний, млрд. руб.\n', fontsize=12)
plt.bar(industry.index, industry.v)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12);
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(14, 8))
plt.title('''\nРаспределение отраслей по выручке за 2017 год ,без учета нефтегазовой отрасли (в рублях)\n''', fontsize=16)
plt.bar(industry[1:].index, industry[1:].v2017)
plt.ylabel('\nВыручка компаний, млрд. руб.\n', fontsize=12)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12);
```

```python
gb = 13488
iog = 5050
pens = 8319

plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(14, 8))
plt.title('''\nРаспределение отраслей по выручке за 2017 год (в рублях)
в сравнении с некоторыми финанасовыми показателями РФ\n''', fontsize=16)
plt.axhline(gb, linestyle='--', linewidth=1, color='red')
plt.axhline(iog, linestyle='--', linewidth=1, color='red')
plt.axhline(pens, linestyle='--', linewidth=1)
plt.ylabel('\nВыручка компаний, млрд. руб.\n', fontsize=12)
plt.text(8, gb + 800,  'Доходы в бюджет РФ в 2017 году (Закон о бюджете) — 13,488 трлн. руб.', fontsize=12)
plt.text(8, iog + 800,   'Нефтегазовые доходы в бюджет 2017 году — 5,05 трлн. руб.', fontsize=12)
plt.text(8, pens + 800,   'Расходы бюджета ПФР в 2017 году — 8,31 трлн. руб.', fontsize=12)
plt.bar(industry.index, industry.v)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12);
```

```python
plt.figure(figsize=(14, 9))
plt.title('\nРаспределение отраслей по выручке за 2013-2017 год (в рублях)\n', fontsize=16)
plt.plot(industry.v2017)
plt.plot(industry.v2016)
plt.plot(industry.v2015)
plt.plot(industry.v2014)
plt.plot(industry.v2013)
plt.ylabel('\nВыручка компаний, млрд. руб.\n', fontsize=12)
plt.legend(fontsize=14)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12);
```

```python
print('Распределение отраслей по совокупной выручке за 2013- 2017 годы (в рублях)')
print('Приведены сравнительные данные по 2017 году')
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(14, 8))
plt.title('\nРаспределение отраслей по совокупной выручке за 2013- 2017 годы (в рублях)\n', fontsize=16)
plt.ylabel('\nВыручка компаний, млрд. руб.\n', fontsize=12)
plt.bar(industry.index, industry.v2017 + industry.v2016 + industry.v2015 + industry.v2014 + industry.v2013)
plt.bar(industry.index, industry.v2017)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12);
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(16, 8))
plt.ylabel('Выручка, млн. руб.\n', fontsize=12)
plt.title('\nРаспределение отраслей по совокупной выручке за 2013-2017 годы (в рублях)\n', fontsize=16)
plt.bar(industry.index, industry.v2013+industry.v2014+industry.v2015+industry.v2016+industry.v2017)
plt.bar(industry.index, industry.v2013+industry.v2014+industry.v2015+industry.v2016)
plt.bar(industry.index, industry.v2013+industry.v2014+industry.v2015)
plt.bar(industry.index, industry.v2013+industry.v2014)
plt.bar(industry.index, industry.v2013)
plt.axhline(industry.v2013.max(), linestyle='--', linewidth=1, color='grey', alpha=0.7)
plt.axhline(industry.v2014[0]+industry.v2013[0], linestyle='--', linewidth=1, color='grey', alpha=0.7)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12);
print('По возрастанию от 2013 до 2017 г.')
```
